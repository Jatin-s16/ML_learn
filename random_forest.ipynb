## SK learn digit recognition dataset

```python
import pandas as pd
from sklearn.datasets import load_digits

digits = load_digits()
```

```python
dir(digits)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.gray()

for i in range(4):
    plt.matshow(digits.images[i])
```

```python
digits.data[:5]
```

## Image size

The images above are basically 8x8 size, so keeping them in a dataframe format converts them to 64 size 1D matrix.

```python
df = pd.DataFrame(digits.data)
df.head()
```

## Feature addition

Now since we have created a new dataframe which only consists of data parameter from the dataset, we also have to incorporate targets parameter.

```python
df['target'] = digits.target
df.head()
```

## Importing test train data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(['target'], axis = 'columns'), digits.target, test_size = 0.2)
```

```python
print(len(X_train))
print(len(X_test))
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators = 1000)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predict)
print(cm)
```

```python
%matplotlib inline
import seaborn as sns

plt.figure(figsize = (10,7))
sns.heatmap(cm, annot = True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
```
